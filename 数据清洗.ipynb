### 加载数据到变量data中

```python
data = pd.read_csv('lagoudata.csv', sep=',')
```

### 去除有缺失值的行 （没有缺失值）

```python
data = data.dropna()
data.head()
```

### 去除“经验要求”和“学历要求”的“/”； 公司名称中的“拉钩认证企业”

```python
data['经验要求'] = data['经验要求'].str.strip('/').str.strip('经验')
data['学历要求'] = data['学历要求'].str.strip('/')
data['公司名称'] = data['公司名称'].apply(lambda x : re.sub('\s+拉勾认证企业', '',x))
data.head()
```

### 处理薪资
1. 将“-”两端的数值分开
2. 将“k”转换为000并存储为数值类型
3. 取两者平均值

```python
salary = data['薪资区间'].str.lower().str.replace('k', '000').str.replace('0以上', '1').str.split('-').apply(pd.Series)
salary.fillna(50000, inplace=True) # 数据中有“20k以上” 以上限50k作为工资区间
data['salary'] = (salary[0].astype('int')+salary[1].astype('int'))/2
```

### 将获取到的具体的工作地址标注在地图上

```python
# 获取工作地址
workPlace = data['工作地址'].apply(lambda x : re.sub('\s+','',x)).str.strip('查看地图')
workPlaceDF = pd.DataFrame(workPlace)
workPlaceDF
```

### 利用百度地图api将文本地址转化为经纬度

```python
import json
from urllib.request import urlopen, quote
import requests
def getlnglat(address):
    url = 'http://api.map.baidu.com/geocoder/v2/'
    output = 'json'
    ak = 'VzKN1NWYidtfuD08x1rwwCtLHguIoZ0f' # 百度地图ak
    address = quote(address) # 由于本文地址变量为中文，为防止乱码，先用quote进行编码
    uri = url + '?' + 'address=' + address  + '&output=' + output + '&ak=' + ak 
    req = urlopen(uri)
    res = req.read().decode() 
    temp = json.loads(res)
    lat = temp['result']['location']['lat']
    lng = temp['result']['location']['lng']
    return lat,lng   # 纬度 latitude   ，   经度 longitude  ，
```

```python
# 将经纬度和工作地址保存在一个DataFrame中
for indexs in data.index:
    get_location = getlnglat(workPlaceDF.loc[indexs,'工作地址'])
    get_lat = get_location[0]
    get_lng = get_location[1]
    workPlaceDF.loc[indexs,'纬度'] = get_lat
    workPlaceDF.loc[indexs,'经度'] = get_lng
```

```python
workPlaceDF.head()
```

```python
# 将“公司名称”和经纬度数据一起写入到一个Series中，以字典的形式
data_html2 = pd.Series()
for indexs in data.index:
    data_html2.loc[indexs] = '{' + \
                                      '"lat":' + str(workPlaceDF.loc[indexs,'纬度']) + ',' +  \
                                      '"lng":' + str(workPlaceDF.loc[indexs,'经度']) + ',' +  \
                                      '"compName":' + '"' + str(data.loc[indexs,'公司名称']) +'"' +   \
                                      '}' + ','

data_html2.to_csv ("data_html_2.csv",encoding="utf-8")
```
